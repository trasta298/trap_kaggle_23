```python
import pandas as pd

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

tmp = [0] * len(train)

animes = dict(zip(train['anime_id'], tmp))
animesCnt = dict(zip(train['anime_id'], tmp))
people = dict(zip(train['user'], tmp))
peopleCnt = dict(zip(train['user'], tmp))

for i in range(len(train)):
    animes[train['anime_id'][i]] += train['score'][i]
    animesCnt[train['anime_id'][i]] += 1
    people[train['user'][i]] += train['score'][i]
    peopleCnt[train['user'][i]] += 1

# anime
animeAvg = {}
for i in animes:
    animeAvg[i] = animes[i] / animesCnt[i]

# user
userAvg = {}
for i in people:
    userAvg[i] = people[i] / peopleCnt[i]

print(animeAvg)

print(userAvg['df5ce90323'])
print(animeAvg['2191'])
```
